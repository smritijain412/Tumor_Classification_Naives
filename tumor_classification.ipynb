```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("data.csv")
data.shape
```

```python
data.head(5)
```

```python
data1=data.drop(labels=['id','Unnamed: 32'],axis=1,inplace=False)
```

```python
data1.head(5)
```

```python
data1.head(5)
```

```python
#change category to numerical
data1['diagnosis'].replace(to_replace=['B','M'],value=[0,1],inplace=True)
```

```python
data1.corr()
```

```python
from sklearn.tree import ExtraTreeClassifier
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data1.iloc[:,1:-1]
Y=data1.iloc[:,0]

```

```python
model = ExtraTreeClassifier()
model.fit(X,Y)
#print(model.feature_importances_) #use inbuilt class feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
X1=data1[['concavity_worst','area_worst','radius_mean','concave points_mean','perimeter_worst']]
```

```python
X1
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
x_train,x_test,y_train,y_test=train_test_split(X1,Y)
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
gaussian=GaussianNB()
gaussian.fit(x_train,y_train)
```

```python
y_pred=gaussian.predict(x_test)
```

```python
### Check Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,y_pred)
```

```python
score
```

```python

### Create a Pickle file using serialization 
import pickle
pickle_out = open("classifier.pkl","wb")
pickle.dump(gaussian, pickle_out)
pickle_out.close()
```
