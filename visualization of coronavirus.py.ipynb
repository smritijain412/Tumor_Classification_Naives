```python
import numpy as np
import pandas as pd
import folium
```

```python
ts=pd.read_csv('time_series_covid_19_confirmed.csv')
```

```python
ts.dropna(axis=0,inplace=True)
```

```python
ts.head()
```

```python
a=ts.iloc[:,4:-1]
```

```python
a
```

```python
total=a.sum(axis=1,skipna=True)
```

```python
total
```

```python
ts['total']=total
```

```python
ts
```

```python
ts.sort_values(by=['total'])
```

```python
ts.dropna(axis=1)
```

```python
location = ts['Lat'].mean(), ts['Long'].mean()
m = folium.Map(location=location,zoom_start=15)

for i in range(0,len(ts)):
       
    popup = folium.Popup('Cornavirus', parse_html=True) 
    folium.Marker([ts['Lat'].iloc[i],ts['Long'].iloc[i]],popup=str(ts.iloc[i]['Country/Region'])+" " + str(ts.iloc[i]['total'])).add_to(m)
m
```

```python
data_heat = ts[['Lat','Long','total']].values.tolist()
```

```python
import folium.plugins as plugins

m = folium.Map(location=location, zoom_start=15)
tiles='stamentoner'
for i in range(0,len(ts)):
       
    popup = folium.Popup('Cornavirus', parse_html=True) 
    folium.Marker([ts['Lat'].iloc[i],ts['Long'].iloc[i]],popup=str(ts.iloc[i]['Country/Region'])+"  "+str(ts.iloc[i]['Province/State'])+str(ts.iloc[i]['total'])).add_to(m)


plugins.HeatMap(data_heat).add_to(m)

m
```
